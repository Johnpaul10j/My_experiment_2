# Jean pockets

Are women jean pockets ridiculously small? Let's look at the data!

The file `jean-pocket-measurements.csv` shows pocket measurements for 20 popular brands.
Four pairs of jeans from each brand were measured: men's and women's skinny and straight styles. All jeans were designated a 32-inch waistband.

The following measurements are included:
- `height_front`: maximum front pocket height (cm)
- `height_back`: maximum back pocket height (cm)
- `width_front`: maximum front pocket width (cm)
- `width_back`: maximum back pocket width (cm)

```python
import pandas as pd
df = pd.read_csv('jean-pocket-measurements.csv')
df
```

### Project Ideas

- What is the average difference in pocket `height_front` between women's and men's jeans?

- Is there a significant difference in pocket `height_front` between skinny and straight styles within the same gender?

- How do back pocket sizes compare between women's and men's jeans?

- Measure your phone's height. What percentage of women's and men's jeans can comfortably fit your phone in the pockets?



```python
# Calculate the average front pocket height for each gender
average_height_by_gender = df.groupby('gender')['height_front'].mean()

# Calculate the difference in average front pocket height
difference = average_height_by_gender['men'] - average_height_by_gender['women']

print(f"The average difference in front pocket height between men's and women's jeans is: {difference:.2f} cm")
```

Is there a significant difference in pocket height_front between skinny and straight styles within the same gender?

```python
# Calculate the average front pocket height for each gender and style
average_height_by_gender_style = df.groupby(['gender', 'style'])['height_front'].mean()

print("Average front pocket height (cm) by gender and style:")
print(average_height_by_gender_style)
```

How do back pocket sizes compare between women's and men's jeans?

```python
# Calculate the average back pocket height and width for each gender
average_back_pocket_by_gender = df.groupby('gender')[['height_back', 'width_back']].mean()

print("Average back pocket size (cm) by gender:\n ")
print(average_back_pocket_by_gender)
```

Measure your phone's height. What percentage of women's and men's jeans can comfortably fit your phone in the pockets?

```python
# Please enter your phone's height in centimeters
phone_height = float(input("Enter your phone's height in cm: "))

# Filter for women's and men's jeans
womens_jeans = df[df['gender'] == 'women']
mens_jeans = df[df['gender'] == 'men']

# Calculate the percentage of women's and men's jeans that can fit the phone in the front pocket
womens_fit_percentage_front = (womens_jeans['height_front'] >= phone_height).mean() * 100
mens_fit_percentage_front = (mens_jeans['height_front'] >= phone_height).mean() * 100

# Calculate the percentage of women's and men's jeans that can fit the phone in the back pocket
womens_fit_percentage_back = (womens_jeans['height_back'] >= phone_height).mean() * 100
mens_fit_percentage_back = (mens_jeans['height_back'] >= phone_height).mean() * 100


print(f"\nPercentage of women's jeans that can fit a {phone_height:.2f} cm phone:")
print(f"- Front pocket: {womens_fit_percentage_front:.2f}%")
print(f"- Back pocket: {womens_fit_percentage_back:.2f}%")

print(f"\nPercentage of men's jeans that can fit a {phone_height:.2f} cm phone:")
print(f"- Front pocket: {mens_fit_percentage_front:.2f}%")
print(f"- Back pocket: {mens_fit_percentage_back:.2f}%")
```
