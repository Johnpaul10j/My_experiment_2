### Project Ideas

- How many territories show Friday, Saturday, Sunday, and Monday as the `first_day` of the week?

- How many people start the week on Friday, Saturday, Sunday, and Monday?
	- Hint: This will involve a `merge`.

- Which of the `four_regions` predominantly start the week on Sunday? On Monday? Are there any regions that are more divided between Sunday and Monday?
	- Hint: This will also involve a `merge`.

```python
first_day_counts = df['first_day'].value_counts()
print("Number of territories with each first day of the week:\n ")
print(first_day_counts[['fri', 'sat', 'sun', 'mon']])
```

How many people start the week on Friday, Saturday, Sunday, and Monday?

```python
merged_df = pd.merge(df, pop, on='alpha3')
population_by_first_day = merged_df.groupby('first_day')['population'].sum()
print("Total population starting the week on each day (in millions):\n ")
print(population_by_first_day[['fri', 'sat', 'sun', 'mon']])
```

Which of the four_regions predominantly start the week on Sunday? On Monday? Are there any regions that are more divided between Sunday and Monday?

```python
# Merge with regions data
merged_regions_df = pd.merge(merged_df, regions, on='alpha3')

# Group by region and first day, then sum population
region_first_day_population = merged_regions_df.groupby(['four_regions', 'first_day'])['population'].sum().unstack(fill_value=0)

# Calculate percentage for Sunday and Monday within each region
region_first_day_population['Total'] = region_first_day_population[['sun', 'mon']].sum(axis=1)
region_first_day_population['Sunday_Percentage'] = (region_first_day_population['sun'] / region_first_day_population['Total']) * 100
region_first_day_population['Monday_Percentage'] = (region_first_day_population['mon'] / region_first_day_population['Total']) * 100

print("Population distribution by region and first day of the week (in millions) and percentages for Sunday and Monday:\n")
print(region_first_day_population[['sun', 'mon', 'Sunday_Percentage', 'Monday_Percentage']])

# Determine predominant day and divided regions
print("\nAnalysis of predominant first day by region:")
for index, row in region_first_day_population.iterrows():
    region = index
    if row['Sunday_Percentage'] > row['Monday_Percentage'] and row['Sunday_Percentage'] > 60: # Using 60% as a threshold for predominant
        print(f"- {region}: Predominantly Sunday")
    elif row['Monday_Percentage'] > row['Sunday_Percentage'] and row['Monday_Percentage'] > 60: # Using 60% as a threshold for predominant
        print(f"- {region}: Predominantly Monday")
    else:
        print(f"- {region}: More divided between Sunday and Monday")
```
