```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('largest-islands.csv')
df.head()
```

### Project Ideas

- What are the 10 largest islands in the tropics?

- What are the largest islands in each `region`?

- Create a line graph with `area` on the y-axis and `rank` on the x-axis. The data should be ordered by `rank`, from largest to smallest.

- What islands are composed of multiple countries?
	- Hint: try `countries.str.contains(",")`

```python
tropics_islands = df[df['climate'] == 'tropics']
largest_tropics_islands = tropics_islands.sort_values(by='area', ascending=False).head(10)
display(largest_tropics_islands)
```

What are the largest islands in each region?

```python
largest_islands_by_region = df.loc[df.groupby('region')['area'].idxmax()]
display(largest_islands_by_region)
```

Create a line graph with area on the y-axis and rank on the x-axis. The data should be ordered by rank, from largest to smallest.

```python
df_sorted = df.sort_values(by='rank', ascending=True)

plt.figure(figsize=(10, 6))
plt.plot(df_sorted['rank'], df_sorted['area'])
plt.xlabel('Rank')
plt.ylabel('Area (km2)')
plt.title('Island Area by Rank')
plt.grid(True)
plt.show()
```

What islands are composed of multiple countries?

```python
multiple_countries_islands = df[df['countries'].str.contains(",")]
display(multiple_countries_islands)
```
